# Seals detection
## Objective
This project aims to create a system that detects seals within an image vy generating a bounding box that covers the desired seal and then crop the bounding box with the seal.


## Process

<b>Images preparation<b>
- Our dataset consisted of images of three different animals: Seals, Cattle, Pig
- Next, we annotated the images, generating one text file per image which consisted of one row with the class and the bounding box coordinates(five values in total).
- We preproccessed the dataset to bring it to a Yolov5 appropriate format.

<b>Yolov5 configuration<b>
- Next step  was to clone the yolov5 repo with the next commands on the terminal
<br>
<b> git clone https://github.com/ultralytics/yolov5 <b>
<br>
<b> cd yolov5<b>
<br>
<b> pip install -r requirements.txt<b>
<br>

<b>Configuration<b>
- Then we have to create a data config file that contains information about the directories of train/test datasets, the number of classes and the names of classes. Just create on the terminal an animal.yaml file on the yolov5/data path.

<b>Training Yolov5<b>
- For training the model, run the next command on the terminal<br>
<b>python train.py --img 640 --batch 16 --epochs 5 --data dataset.yaml --weights yolov5s.pt<b><br>

With the above command we tell python to run the train file, on x number of epochs and batches, with the pretrained weights from the COCO dataset. Those weights are not very relevant for our dataset as they are not trained on seals sperccifically, but they are still better than random initialization.
    
<b>Prediction/Detection<b><br>
    
After training is done, the best epoch weighs will be saved in a directory like yolov5/runs/train/exp/weights/best.pt.
For testing the model on new untested images, we can run the next command<br>
- <b>python detect.py --weights /runs/train/exp5/weights/best.pt --conf 0.1 --source [ test image directory ]<b><br>

The conf argument is to filter bounding boxes under the confidence of 0.1. <br>
After the detections run, the predictions will be saved in a new directory which will be displayed like below 

    
For getting the cropped image, just run the following command, and the results will be saved on the same manner.
- <b>python detect.py --weights /runs/train/exp5/weights/best.pt --conf 0.1 --source [ test image directory ] --crop-save<b><br>
